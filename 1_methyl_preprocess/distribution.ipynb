```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Adjust delimiter if needed (default is tab for BED)
bed = pd.read_csv("/home/alsun/get/1_methyl_preprocess/SAA-125-2M_L2-3.CGN-both.frac.filtered.bed", sep='\t', header=None)

# View first few lines
bed.head()

# Use seaborn for a smooth distribution
sns.histplot(bed[3], bins=50, kde=True)

# Customize plot
plt.title("Distribution of Accessibility Scores")
plt.xlabel("Accessibility Score")
plt.ylabel("Frequency")
plt.show()
```
