```python
import pandas as pd
import matplotlib.pyplot as plt

# Load BED file assuming it's tab-separated and has at least 3 columns: chrom, start, end
bed = pd.read_csv("/home/alsun/get/1_methyl_pre_070725/L2_3_IT_CTX_Glut.CGN.bed", sep="\t", header=None, usecols=[1, 2], names=["start", "end"])
bed["size"] = bed["end"] - bed["start"]

# Plot histogram
plt.hist(bed["size"], bins=50, color='steelblue', edgecolor='black')
plt.xlim(0, 2000)
plt.title("Distribution of DMR Sizes For L2/3")
plt.xlabel("DMR Size (bp)")
plt.ylabel("Frequency")
plt.show()
```
